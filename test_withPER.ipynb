```python
import pandas as pd
import numpy as np
from sklearn import mixture
from sklearn.metrics import accuracy_score
import pickle
import os
from os import path
import sys
import glob
```

```python
# Need to open the pickled list object into read mode
n_comp = 16
# list_pickle_path = './models/mfcc'+str(n_comp)+'.pkl'
# list_unpickle = open(list_pickle_path, 'rb')

# load the unpickle object into a variable
# weighted_probs = pickle.load(list_unpickle)
# print(weighted_probs)

df = pd.read_hdf("./features_test/mfcc/timit.hdf")
print(df.head())
features = np.array(df["features"].tolist())
features = np.round(features,3)
labels = np.array(df["labels"].tolist())
```

```python
l=[]
list_pickle_path = './models/mfcc/'+str(n_comp)+'/'
# list_unpickle = open(file_path, 'rb')
# print(list_unpickle)
# gmm = pickle.load(list_unpickle)

labels_order = []
labels_score_array = []
for filename in os.listdir(list_pickle_path):
	if(filename.endswith(".pkl")):
		os.path.join(filename,list_pickle_path)
		file_path = list_pickle_path+str(filename)
		list_unpickle = open(file_path, 'rb')
		gmm = pickle.load(list_unpickle)
		score = gmm.score_samples(features)
		labels_order.append(filename[:-4])
		labels_score_array.append(score)
```

```python
labels_num = []
for i in range(len(labels)):
	labels_num.append(labels_order.index(labels[i]))
labels_score_array1 = np.column_stack(labels_score_array)
predicted_labels_num=np.argmax(labels_score_array1,axis=1)
predicted_labels_num=np.array(predicted_labels_num)
labels_num = np.array(labels_num)
```

```python
# labels_score_array = np.array(labels_score_array)
acs = accuracy_score(labels_num,predicted_labels_num)
print(acs)
```

```python
equal =0
count=0
frame_ac=[]
for i in range(len(labels_num)):
    count=count+1
    if(labels_num[i]==predicted_labels_num[i]):
        equal = equal + 1
    if(i==len(labels_num)-1 or labels_num[i]!=labels_num[i+1] ):
        frame_ac.append((equal*100)/count)
        equal=0
        count=0
    
print(frame_ac)
```
